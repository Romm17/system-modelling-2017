```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from scipy.stats import chisquare
```

# Some useful stuff

```python
def plot_dist(ax, dist, title, color):
    mean = np.mean(dist)
    std = np.std(dist)
    ax.set_title("{:10}, mean = {:.3f}, std = {:.3f}".format(title, mean, std))
    ax.hist(dist, color=color)
```

# Generating distributions

```python
size = 10000

uniform = np.random.uniform(size=size)
normal = np.random.normal(size=size)
exponential = np.random.exponential(size=size)
```

# Ploting results

```python
f, (ax1, ax2, ax3) = plt.subplots(3, figsize=(10, 9), sharex=True)
plot_dist(ax1, uniform, "Uniform", 'r')
plot_dist(ax2, normal, "Normal", 'g')
plot_dist(ax3, exponential, "Exponential", 'b')
f.subplots_adjust(hspace=0.3)
plt.show()
```

# Test $ \chi^2 $

```python
print(chisquare(uniform, uniform))
print(chisquare(normal, normal))
print(chisquare(exponential, exponential))
```
